```python
# oading the dataset from the huggin face library 
from datasets import load_dataset

ds = load_dataset("SandipPalit/Movie_Dataset")
```

```python
# turn the dataset into a pandas dataframe
df = ds["train"].to_pandas()
```

```python
# only keep records with Genre containing "Horror"
df3 = df[df["Genre"].str.contains("Horror")]
```

```python
df5 = df[df["Overview"].str.contains("ghost")]
```
